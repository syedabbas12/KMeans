```python
from sklearn import datasets
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data
df = pd.DataFrame(X)
```

```python
df.columns = iris.feature_names
df
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
%matplotlib inline
scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
```

```python
def find_k (X, k_range, sample_percent=1):
    
    from sklearn.cluster import KMeans
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.spatial.distance import cdist
    from sklearn.metrics import pairwise_distances

    N = X.shape[0]
    sampleSize = X.shape[0] * sample_percent

    if sampleSize > 0:
        index = np.arange(np.shape(X)[0])
        np.random.shuffle(index)
        X =  X[index, :]


    mean_distortions=[]
    for k in k_range:
        kmeans_model = KMeans(n_clusters=k, init='k-means++', n_jobs=-1).fit(X)
        mean_distortions.append(sum(np.min(pairwise_distances(X, kmeans_model.cluster_centers_,metric='euclidean'),axis=1)) / X.shape[0])
    plt.plot(k_range, mean_distortions)
    plt.xlabel("K Value")
    plt.ylabel("Mean Distortion")
    plt.show()
```

```python
find_k(np.matrix(X), range(1, 10), 1)
```

```python
kmeans_model= KMeans(n_clusters=2, random_state=42)
kmeans_model.fit(X_scaled)
```

```python
df["lables"]= kmeans_model.labels_
df.head(10)
```
